## .nunique() method in Python Pandas

The `.nunique()` method in Python Pandas is used to count the number of unique values in a Series or DataFrame. It takes an optional argument called `axis` which specifies the axis over which the uniqueness is to be determined. By default, the axis is 0, which means that the number of unique values in each column is returned. If axis is 1, then the number of unique values in each row is returned.

The `.nunique()` method also has an optional argument called `dropna` which specifies whether or not to exclude NaN values from the count. By default, NaN values are excluded.

**Example:**

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 1, 2, 2, 3, 3, 3], 'B': ['a', 'b', 'a', 'b', 'c', 'c', 'd']})

print(df['A'].nunique())
# Output: 3

print(df['B'].nunique())
# Output: 4


**Explanation:**

In the above example, the `.nunique()` method is used to count the number of unique values in the `A` and `B` columns of the DataFrame `df`. The output shows that there are 3 unique values in the `A` column and 4 unique values in the `B` column.

**Additional notes:**

* The `.nunique()` method can be used on both Series and DataFrames.
* The `.nunique()` method is not affected by the order of the values in a Series or DataFrame.
* The `.nunique()` method is a very efficient method and it can be used to count the number of unique values in large datasets.

## .count() method in Python

The `.count()` method in Python is used to count the number of times a specific value appears in a list or string. It takes an optional argument called `value` which specifies the value to be counted. If the `value` argument is not specified, then the method will count the number of occurrences of all the elements in the list or string.

**Example:**

```python
# Create a list
list = [1, 1, 2, 2, 3, 3, 3]

# Count the number of occurrences of 3
print(list.count(3))
# Output: 3


In the above example, the `.count()` method is used to count the number of occurrences of the number 3 in the list `list`. The output shows that the number 3 appears 3 times in the list.

**Explanation:**

The `.count()` method works by iterating over the list or string and checking if each element is equal to the value specified in the `value` argument. If the element is equal to the value, then the method increments a counter. The method returns the value of the counter.

**Additional notes:**

* The `.count()` method can be used on both lists and strings.
* The `.count()` method is not affected by the order of the elements in a list or string.
* The `.count()` method is a very efficient method and it can be used to count the number of occurrences of a value in large datasets.

## .value_counts() method in Python Pandas

The `.value_counts()` method in Python Pandas is used to count the number of occurrences of each unique value in a Series or DataFrame. It returns a Series object with the unique values as the index and the counts as the values.

**Example:**

```python
import pandas as pd

# Create a Series
series = pd.Series([1, 1, 2, 2, 3, 3, 3])

# Count the number of occurrences of each unique value
value_counts = series.value_counts()

# Print the value_counts
print(value_counts)

# Output:
# 3    3
# 1    2
# 2    2


**Explanation:**

The `.value_counts()` method works by iterating over the Series or DataFrame and counting the number of times each unique value appears. The results are stored in a Series object with the unique values as the index and the counts as the values.

**Additional notes:**

* The `.value_counts()` method can be used on both Series and DataFrames.
* The `.value_counts()` method is not affected by the order of the values in a Series or DataFrame.
* The `.value_counts()` method can be used to count the number of occurrences of missing values (NaN) by setting the `dropna` argument to False.

## .info() method in Python Pandas

The `.info()` method in Python Pandas is used to get a summary of the statistical information about a Series or DataFrame. It includes the number of rows, columns, data types, memory usage, and descriptive statistics for each column.

**Example:**

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 1, 2, 2, 3, 3, 3], 'B': ['a', 'b', 'a', 'b', 'c', 'c', 'd']})

# Get the information about the DataFrame
print(df.info())

# Output:
# <class 'pandas.core.frame.DataFrame'>
# RangeIndex: 7 entries, 0 to 6
# Data columns (total 2):
#   A    int64
#   B    object
# dtypes: int64(1), object(1)
# memory usage: 25.6 KB
# None



**Explanation:**

The `.info()` method works by printing a summary of the DataFrame, including the following information:

* The class of the DataFrame
* The number of rows and columns
* The names of the columns
* The data types of the columns
* The memory usage of the DataFrame
* Descriptive statistics for each column, such as the mean, standard deviation, and minimum and maximum values

**Additional notes:**

* The `.info()` method can be used on both Series and DataFrames.
* The `.info()` method does not modify the original DataFrame.
* The `.info()` method can be used to get a summary of the statistical information about a large dataset.
